```python
def proc_images():
    """
    Returns two arrays: 
        x is an array of resized images
        y is an array of labels
    """
    
#     disease="Infiltration"

    x = [] # images as arrays
    y = [] # labels Infiltration or Not_infiltration
    WIDTH = 224
    HEIGHT = 224
    counter = 1
    total = len(images)
    for img in images:
        if counter % 1000 == 0:
            print('resized image {} of {}'.format(counter,total)
#         print('=============')
        base = os.path.basename(img)
        print(base)
        name = base.split(".")[0]
        # Read and resize image
        full_size_image = cv2.imread(img)
        resize_img = cv2.resize(full_size_image, (WIDTH,HEIGHT), interpolation=cv2.INTER_CUBIC)
        
        # write the cropped image to disk in PNG format
        cv2.imwrite('sample_resize/'+name +".png",resize_img)
#         print('done with picture ')
        counter +=1
```
